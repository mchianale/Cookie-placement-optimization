# Random method with scoring

In this algorithm, we will just give a score for each biscuit between 0 and 1 in.

More the score of a biscuit close to one, the more the probability to place this biscuit in the roll increases.

 ## How to give scores to each biscuit?

- Create a valuable benefit for each biscuit: its value / its leng.th
- Take in consideration the number of defects which can have biscuits, more a biscuit can have a defect more its score increas
- normalized the features of biscuits to have a score between 0 and 1.
- apply to each feature a coefficient : importance rate, the sum of importance rates must be equal to 1.
- The final score is the sum of each feature value * importance rate.
- Finally, to deduce the probabilitie of each biscuit we will normalized final scores.o 1

### features to take in consideration

```python
values = np.array([6,12,1,8])
lengths = np.array([4,8,2,5])
a = [4,5,1,2]
b = [2,4,2,3]
c = [3,4,1,2] 

df_biscuit = pd.DataFrame({
    'benefit' : values/lengths,
    'a' : a,
    'b' : b,
    'c' : c })

df_biscuit.head()
```

### Normalized the features

```python
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())
# Normalize each column
df_biscuit = df_biscuit.apply(min_max_normalize)
df_biscuit.head()
```

### Apply importance rates to the features

```python
importance_rates = [0.4, 0.2, 0.2, 0.2] #sum musts be equal to one
for i, col in enumerate(df_biscuit.columns):
    df_biscuit[col] = df_biscuit[col]*importance_rates[i]
df_biscuit.head()
```

### Give final scores

```python
df_biscuit['score'] = df_biscuit.sum(axis=1) + 0.0001 #avoid to have after null probabilities
df_biscuit.head()
```

```python
scores = df_biscuit['score']
#add a eps value to score to prevent to have a probabilitie equals to zero !
df_biscuit['probabilitie'] =  [(score+0.0001) / sum(scores) for score in scores ] #sum musts be equal to one
df_biscuit.head()
```

## Take scoring and probabilities in consideration:

```python
def random_scoring_roll_rec_(self, scores, start=0, max_iter=100):
        """ place biscuit randomly in the roll but in depends of probabilities """
        #remain space is to small to place any existing biscuit
        if len(self.biscuits) - start < MINLENGTH:
                return 
        #compute probabilities
        scores_ = scores.copy()
        probabilities = [(score) / sum(scores_) for score in scores_] 
    
        n_possibilities = len(BISCUITS)
        possible_biscuits = BISCUITS.copy()
        #pick a biscuit randomly and consider probabilities
        index = np.random.choice(n_possibilities, size=1, p=probabilities)[0]
        biscuit = possible_biscuits[index].copy()
        already_tried = []
        
        while not self.addBiscuit(biscuit, start):
            #update current possibilities
            already_tried.append(biscuit)
            #no possibilities at this position, start from second pos from current start
            if len(already_tried) == 4:
                return self.random_scoring_roll_rec_(scores,start+1)
                
            #update current possibilities
            n_possibilities -= 1
            possible_biscuits.pop(index)
                
            #recomputes probabilities
            scores_.pop(index)
            probabilities = [score / sum(scores_) for score in scores_]

            #choose an another posibilitie
            index = np.random.choice(n_possibilities, size=1, p=probabilities)[0]
            biscuit = possible_biscuits[index]
            
        return self.random_scoring_roll_rec_(scores, biscuit['length']+start)
    
Roll.random_scoring_roll_rec_ = random_scoring_roll_rec_
```

## Variation of scoring

Now, we can apply different scores:

- In our case, we can just consider more benefits in the score than constraints.
- But it is also interesting to consider constraints, because if you run several times the algorithm, you will gain time by increasing probabilities of biscuits which can have more constraints than others.

```python
values = np.array([6,12,1,8])
lengths = np.array([4,8,2,5])
a = [4,5,1,2]
b = [2,4,2,3]
c = [3,4,1,2] 
FEATURES = pd.DataFrame({
    'benefit' : values/lengths,
    'a' : a,
    'b' : b,
    'c' : c })
#normalized
FEATURES = FEATURES.apply(min_max_normalize)
FEATURES.head()
```

```python
def getScore(importance_rates, eps=0.001):
    """ return array of probs in depends of importance_rates input """
    f = FEATURES.copy() 
    #apply importance rate
    for i, col in enumerate(f.columns):
        f[col] = f[col]*importance_rates[i]
    scores = list(f.sum(axis=1)+eps)
    return scores
```

First analysis, we will make variation of importance rate for the benefit, starting at 0.1 to 1. 

Constraints rates will just be equal to what is staying.

```python
all_importance_rates = [ [i, (1-i)/3, (1-i)/3, (1-i)/3] for i in np.arange(0.1,1.1, 0.1)]
```

```python
import matplotlib.pyplot as plt
```

```python
max_iter = 1000

mean_values = []
max_values = []
mean_times = []
benefit_rates = []
for importance_rates in all_importance_rates:
    scores = getScore(importance_rates)
    values = []
    times = []
    for i in range(1, max_iter+1):
            main_roll.reset()
            display(HTML(f'<h2>Current Iteration: {i + 1}</h2>'))
            start = time.time()
            main_roll.random_scoring_roll_rec_(scores)
            end = time.time()
            times.append(end-start)
            values.append(main_roll.value) 
            clear_output(wait=True)
        
    mean_values.append(np.array(values).mean())
    max_values.append(np.array(values).max())
    mean_times.append(np.array(times).mean())
    benefit_rates.append(importance_rates[0])
```

## Using Partitions to define constraint rates

### Global Vision

In previous part, we defined for each constraint a rate equals to the remain value of (1-benefit rate)/3, so we had add the same importance for each constraint.

But we can changed these rates by looking on the total number of each constraint.

For example, if on the roll we have 100 defects 'a' and 25 of 'b' and 'c', we will more consider importance rate of 'a'.

```python
main_roll.getNumbersDefects()
```

```python
def auto_constraint_rates(self, benefit_rate):
    remain = 1 - benefit_rate
    n_defects = self.getNumbersDefects()
    a = n_defects['a']
    b = n_defects['b']
    c = n_defects['c']
    total = a + b + c
    if total != 0:
        a_rate = (remain*a)/total
        b_rate = (remain*b)/total
        c_rate = (remain*c)/total
        return [benefit_rate, a_rate, b_rate, c_rate]
    return [benefit_rate, 0, 0, 0]
Roll.auto_constraint_rates = auto_constraint_rates
```

### Main Function

```python
def random_partitions_probabilities(self, n_parts, min_size, benefit_rate):
    #create the partitions 
    divisions = self.divide_roll_randomly(n_parts, min_size)
    partitions = []
    for div in divisions:
        part = {'partition' : self.partition(div[0],div[1]-1), 'start' : div[0]}
        partitions.append(part)
    #for each partitions apply the algorithm and update the roll
    for part in partitions:
        part['partition'].reset()
        #compute each probabilities for each partition
        importance_rates = part['partition'].auto_constraint_rates(benefit_rate)
        scores = getScore(importance_rates)
        part['partition'].random_scoring_roll_rec_(scores)
        #update the roll
        self.updateFromPartition(part['partition'], part['start'])
        
Roll.random_partitions_probabilities = random_partitions_probabilities    
```

# Artificial Bee Algorithm, recursive version

- Now we want to apply previous algorithm but recursively
- To made this we start abc on the main roll.
- The idea is to recall abc on smaller roll (partitions of main_roll).
- Give the possibilitie to go back when you reach the smallest partitions or when you don't make improvement.
- When you apply abc on smaller partitions, limit value needs to decrease because you have less possibilities of biscuit's placements.
- minimum limit is set to 5, we had good results with this value.
- To avoid unnecessaries runs of abc algorithms, we added an attribut max_update, so when no improvements are made we decrease max_update, and if it reachs 0 we stop the algorithm.

## The main function

```python
MIN_LENGTH = 2 #length of the smallest biscuit
def abc_rec(roll,k, max_update):
    #allows to not run the algorithm when we don't find anymore improvement
    if max_update == 0:
        return 
    if roll.size < MIN_LENGTH: #we now we can't place biscuit on a roll of size 1
        #we go back by choose randomly k
        return abc_rec(roll, random.randint(1, k-1), max_update)
        
    #more we go deeper, more we created small partitions
    min_length_base = 500//k
    min_length = random.choice([min_length_base + i for i in range(2)]) #max(random.randint(4,16), roll.size//k)
    #choose a partition of current roll to improve with abc
    current_partition_dict = roll.TournamentPartition(min_length)
    current_partition = current_partition_dict['partition']
 
    #compute the limit, limit decreases when we go deeper in the iteration
    limit = max(5, (max_iter//2)//k)
    abc = ABC(current_partition, colony_size, max_iter,limit)
    optimal_roll, optimal_value = abc.optimize()

    #we made an improvement
    if k == 1 or optimal_value > current_partition.value:
        #we update current roll with the new partition
        roll.updateFromPartition(optimal_roll, current_partition_dict['start'])
        return abc_rec(roll, k+1, max_update)
        
    #failed to improve the current partition
    max_update -= 1
    #let's give several possibilitie, in this case, either go back or move forward
    return abc_rec(roll, random.choice([k-1,k+1]), max_update)
        
def run_abc_rec(roll, colony_size = 5, max_iter = 50, max_update=20):
    """ function to manage recursive abc, currently it is not very use full, perhaps it allows us to add more functionnalities later """
    roll.reset()
    abc_rec(roll, 1, max_update)
```
